```python
#This function converts a column in to numerics for better machine understanding
def column_numerizer(column, list_to_change):
    d_dict = {}
    for i in list_to_change:
        d_dict[i] = list_to_change.index(i)
    j = None
    
    for i in range(column.count()):
        if column[i] in list_to_change:
            for k in range(len(list_to_change)):
                if column[i] == list_to_change[k]:
                    j = list_to_change[k]
                    #print(j)
                    column[i] = d_dict[j]
```

```python
#This function converts list in a string to list
def rem_noise(disturbed_list):
    cleaned_list = disturbed_list.replace("[", "")
    cleaned_list = cleaned_list.replace("]", "")
    cleaned_list = cleaned_list.replace(",", "")
    cleaned_list = cleaned_list.replace("'", "")
    cleaned_list = cleaned_list.split()
    return cleaned_list
```

```python
def extract_unique(array_list):
    list_of_uniques = []
    for i in range(len(array_list)):
        array_list[i] = rem_noise(array_list[i])
        for j in array_list[i]:
            if j not in list_of_uniques:
                list_of_uniques.append(j)
    return list_of_uniques
        
```

```python
class network(nn.Module):
    def __init__(self):
        super(network, self).__init__()
        self.fc1 = nn.Linear(len(working_df.columns)-1, 69)
        self.fc2 = nn.Linear(69, 10)
        self.fc3 = nn.Linear(10, 1)
        
        self.dropout = nn.Dropout(0.3)
    
    def forward(self, x):
        x = torch.tensor(x)
        #print(x)
        x = x.view(x.shape[0], -1)
        #x = torch.reshape(x, (-1,))
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        #print(x)
        return x
```
