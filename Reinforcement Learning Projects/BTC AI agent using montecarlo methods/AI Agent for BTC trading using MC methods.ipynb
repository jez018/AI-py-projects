```python
def MDP():
    states, _ = states_creation()
    s_states = []
    actions = []
    rewards = []
    es_episodes = []
    es_actions = []
    es_rewards = []
    count = 0
    for index, row in states.iterrows():
        count += 1
        # You can access each column value in the 'row' variable
        ## states ##actions ##rewards
        s_states.append((tuple(np.array(row))))
        actions.append((0,1,2))
        rewards.append((-1, -10, -5))
        if count == 30:
            es_episodes.append(s_states)
            es_actions.append(actions)
            es_rewards.append(rewards)
            s_states = []
            actions = []
            rewards = []
            count = 0
    return es_episodes, es_actions, es_rewards
```

```python
def get_probs(Q_s, epsilon, nA):
    policy_s = np.ones(nA) * epsilon / nA
    best_a = np.argmax(Q_s)
    policy_s[best_a] = 1 - epsilon + (epsilon / nA)
    return policy_s
#[0.3333333333333333, 0.3333333333333333, 0.3333333333333333]
```

```python
def episodes_gen_based_on_policy(prob = get_probs, MDP = MDP(), Q=None, epsilon=None, nA=3):
    es_episodes, es_actions, es_rewards = MDP
    episodes_all = []
    #wallet_balance = []
    #wallet = VirtualWallet(6000, 'wallet.txt')
    #vee = VirtualExchanger(wallet, 0.04)
    for ep in es_episodes:
        i = 0
        episode = []
        s_states = ep
        #state = s_states[1]
        while True:
            s_reward = None
            state = s_states[i]
            f_state = state[-1] #This stands for the future state
            state = state[:-1] #This is all the state but not including the future state
            #print('es_action: ', es_actions[i][1])
            action = np.random.choice(es_actions[i][1], p=prob(Q[state], epsilon, nA)) if state in s_states else (s_states.append(state), random.choice([0, 1, 2]))[-1]
            #print('action: ', action)
            if action == 0:
                #exchange buy here
                if f_state > 0:
                    s_reward = -1
                elif f_state < 0:
                    s_reward = -10
                elif f_state == 0:
                    s_reward = -5
            if action == 1:
                #exchange hold here
                if f_state == 0:
                    s_reward = 0
                elif f_state < 0:
                    s_reward = -5
                elif f_state > 0:
                    s_reward = -5
            if action == 2:
                #exchange sell here
                if f_state < 0:
                    s_reward = -1
                elif f_state > 0:
                    s_reward = -10
                elif f_state == 0:
                    s_reward = -5
            episode.append((state, action, s_reward))
            i += 1
            if i > len(es_actions)-1:
                break
        episodes_all.append(episode)
    return episodes_all #, wallet_balance 
```

```python
def updateQ(gamma = 0.3, alpha = 0.05, Q = defaultdict(lambda: np.zeros(3)), episodes_all=None):
    for episode in episodes_all:
        #print(episode)
        states, actions, rewards = zip(*episode)
        discounts = np.array([gamma**i for i in range(len(rewards)+1)])
        #This loop updates Q-table to sub-optimal policy
        for i, state in enumerate(states):
            old_Q = Q[state][actions[i]]
            Q[state][actions[i]] = old_Q + alpha*(sum(rewards[i:]*discounts[:-(1+i)])-old_Q)
    return Q
```

```python
import sys
def control(e_duration, alpha, gamma=1.0, eps_start=1, eps_min=0.05, eps_decay=.99999):
    epsilon = eps_start
    Q = defaultdict(lambda: np.zeros(3))
    #wallet_balance = []
    for i_episode in range(1, e_duration+1):
        # monitor progress
        if i_episode % 10 == 0:
            print("\rEpisodes {}/{}.".format(i_episode, e_duration), end="")
            sys.stdout.flush()
            
        epsilon = max(epsilon*eps_decay, eps_min)
        episodes = episodes_gen_based_on_policy(prob = get_probs, MDP = MDP(), Q=Q, epsilon=epsilon, nA=3) #This should return 2 variables: episodes and wallet_balance
        #print(episodes[])
        Q = updateQ(gamma = gamma, alpha = 0.05, Q = Q, episodes_all=episodes)
    policy = dict((k, np.argmax(v)) for k, v in Q.items()) 
    return policy#, wallet_balance #this should return policy and wallet balance
```

```python
class VirtualWallet:
    def __init__(self, init_amount, file_loc):
        self.f_loc = file_loc
        self.i_amt = init_amount
        
        with open(self.f_loc, 'w') as file:
            file.write(str(self.i_amt))
    
    def credit(self, amount):
        init_amt = 0.0
        with open(self.f_loc, 'r') as file:
            init_amt = float(file.read())
        init_amt = init_amt + amount
        with open(self.f_loc, 'w') as file:
            file.write(str(init_amt))
    
    def debit(self, amount):
        init_amt = 0.0
        with open(self.f_loc, 'r') as file:
            init_amt = float(file.read())
        init_amt = init_amt - amount
        with open(self.f_loc, 'w') as file:
            file.write(str(init_amt))
            
    def show_balance(self):
        balance = None
        with open(self.f_loc, 'r') as file:
            balance = float(file.read())
        #print('Balance: ', balance)
        return balance
```
