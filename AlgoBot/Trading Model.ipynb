```python
import torch
from torch import nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#put train and test in batches using data loader
train_loader = DataLoader(torch.cat((torch.tensor(X_train.values.astype(np.float32)), torch.tensor(y_train.values.astype(np.float32))), dim=1), batch_size=32)
test_loader = DataLoader(torch.cat((torch.tensor(X_test.values.astype(np.float32)), torch.tensor(y_test.values.astype(np.float32))), dim=1), batch_size=32)
```

```python
input_len = len(X_train.columns)
input_len
```

```python
#Defining the neural net
class clf(nn.Module):
    def __init__(self, input_len):
        super(clf, self).__init__()
        self.fc1 = nn.Linear(input_len, 10240)
        self.fc2 = nn.Linear(10240, 13660)
        self.fc3 = nn.Linear(13660, 640)
        self.fc4 = nn.Linear(640, 10)
        self.fc5 = nn.Linear(10, 1)
        
        self.dropout = nn.Dropout(0.2)
    def forward(self, x):
        x = torch.tensor(x)
        x = x.view(x.shape[0], -1)
        
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.fc3(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.fc4(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.fc5(x)
        x = F.sigmoid(x)
        
        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

```python
torch.cuda.is_available()
```

```python
#Defining the optimizers and loss functions 
model = clf(input_len)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.BCELoss()
#model.to('cuda')
```

```python
epoch = 500000
t_loss = []
for e in range(epoch):
    running_loss = 0
    for data in train_loader:
        X_train = data[:, :-1]
        y_train = data[:, -1]
        #X_train.to('cuda')
        #y_train.to('cuda')

        optimizer.zero_grad()

        pred = model(X_train)
        pred = pred.view(-1) #converting to same shape as y_train
        #print(pred)
        #print(y_train)
        loss = criterion(pred, y_train)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
    else:
        print('loss: ', (running_loss/len(train_loader)))
```
